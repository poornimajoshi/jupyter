```python
# DATS 6103 - Individual Project 1 - Poornima Joshi

import pandas as pd
import matplotlib.pyplot as plt


raw_data = pd.read_excel("SIPRI-Milex-data-1949-2017.xlsx", sheet_name="Current USD")
raw_data.columns = raw_data.iloc[4]
raw_data.head()
```

```python
raw_data = raw_data.drop([0,1,2,3,4])
raw_data = raw_data.drop(columns = ['Notes'])
raw_data =  raw_data.set_index('Country')
```

```python
country_list = ['USA','China, P.R.', 'Russian Federation', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Israel']
df = raw_data.loc[country_list]
df
```

```python
col_list = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]
df = df[col_list]
#df = df.reset_index()
df
```

```python

for i in df.index:
    plt.plot(df.loc[i])

plt.xlabel('Year')
plt.ylabel('Military spending in USD')
plt.title('Military spending of several countries')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
import plotly.plotly as py
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
df.head(10)
data = [ dict(
        type = 'choropleth',
        locations = df['CODE'],
        z = df['GDP (BILLIONS)'],
        text = df['COUNTRY'],
        colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            tickprefix = '$',
            title = 'GDP<br>Billions US$'),
      ) ]

layout = dict(
    title = '2014 Global GDP<br>Source:\
            <a href="https://www.cia.gov/library/publications/the-world-factbook/fields/2195.html">\
            CIA World Factbook</a>',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='d3-world-map' )
```
