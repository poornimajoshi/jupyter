```python
import pandas as pd
```

```python
filename = 'output.txt'
data = pd.read_csv(filename, sep=" ", header=None)

data.columns = ["hostname", "col2", "col3", "timestamp", "timezone", "request", "request2", "response code", "total_bytes"]
data['time'] = data[['timestamp', 'timezone']].apply(lambda x: ''.join(x), axis=1)
data
```

```python
host_count = data.groupby('hostname')['hostname'].count()
host_count
```

```python
host_count.to_csv('record_'+filename, sep=' ', mode='a')
```

```python
arr = [1,2,3,4,5, 100000000000]
def findNumber(arr, k):
    if arr <1 or arr > 10**9:
        print('NO')
    if k <= 1 or k >= 10**5:
        print('No')
    elif k in arr:
        print('YES')
    else:
        print('NO')

findNumber(arr, 5)
```
