```python
# DATS 6103 - Individual Project 1 - Poornima Joshi

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the data
### Importing data from the SIPRI Excel file

```python
# Importing data from the SIPRI Excel file
# Using Constant USD as of 2016 since,
# Constant dollars are used to adjust for the effects of 
# inflation by converting economic information into a standard era dollar term
raw_data = pd.read_excel("military.xlsx", sheet_name="Constant (2016) USD", skiprows = 5, index_col='Country')
raw_data.head()
```

#### Cleaning the data to retain only the data with the countries and years of our selection

```python
# Cleaning the data to retain only the data with the countries and years of our selection
country_list = ['USA','China, P.R.', 'Russian Federation', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Israel']
col_list = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]

df = raw_data.loc[country_list]
df = df[col_list]
df
```

# Compare the data to the overall military spending of the all 10 countries
#### Understanding the Military spending of the selected countries

```python
mil_sum = df.T.mean()
mil_sum.plot(kind='pie', subplots=True, figsize=(10, 8), autopct='%1.1f%%')
plt.ylabel('Military Spending by country')
```

```python
#A line graph to understand the Military spending of the selected countries
for i in df.index:
    plt.plot(df.loc[i])

plt.ylabel('Military spending in USD')
plt.title('Military spending of several countries')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

#### Zooming in to exclude USA and China to take a closer look at other countries

```python
df_exclude_US_China = df.copy()
df_exclude_US_China = df_exclude_US_China.drop(df_exclude_US_China.index[[0,1]])
for i in df_exclude_US_China.index:
     plt.plot(df.loc[i])

plt.ylabel('Military spending in USD')
plt.title('Military spending of several countries')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

df_exclude_US_China
```

### Over all summary of all year Military spending per country

We can observe that USA has the spending and China follows in next

```python
# Compute Percentages
df1 = df.copy()
df2 = df1.apply(lambda x : x/x.sum() * 100)
# Transpose to get YoY data
df3 = df2.T
#Visualize area chart to check the countries with the biggest budget YoY

ar = df3.plot.area()
ar.set_ylabel('Military spending in USD')
ar.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

#### Observing the increase or decrease in Military spending.

We can see that USA has reduced it's spending in 2017 when compared to 2008.
China however is gradually increasing its military spending over the years.

China has been growing the fastest over the years. This is seen very clearly seen the following graph

```python
# Year wise comparison of the military spending of the countries
colors = ['Red', 'Plum', 'Yellow', 'Green', 'Blue', 'Orange', 'Purple', 'Gold',
          'Magenta', 'Grey']
ax = df3.plot( kind='bar', width= .6, figsize = (25,15), color = colors, fontsize = 16, 
                 title = 'Military spending of Several countries')
plt.legend(loc = 'best')
ax.set_xlabel('Year', fontsize = 22)
ax.set_ylabel('Military Spending [Millions of US Dollars])', fontsize = 20)
```

Excluding USA and China to see other countries better

# Compare the data to that country’s GDP

#### Importing and processing the Share of GDP data from SIPRI

```python
raw_data_gdp = pd.read_csv("gdp.csv", skiprows=4)
df_gdp = raw_data_gdp.drop(columns=['Country Code','Indicator Name','Indicator Code','Unnamed: 62'])
df_gdp = df_gdp.rename(columns={'Country Name': 'Country', '2008': 2008, '2009': 2009,'2010': 2010, '2011': 2011,'2012': 2012,'2013': 2013,'2014': 2014,'2015': 2015,'2016': 2016,'2017': 2017})
df_gdp = df_gdp.set_index('Country')
country_list_gdp = ['United States','China', 'Russian Federation', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'Korea, Rep.', 'Israel']
df_gdp = df_gdp.loc[country_list_gdp]
df_gdp = df_gdp.rename(index={'United States': 'USA', 'China': 'China, P.R.', 'United Kingdom': 'UK','Korea, Rep.': 'Korea, South'})
df_gdp = df_gdp[col_list]
```

```python
# gdp increase of several countries
ax1 = df_gdp.T.plot(alpha=0.75, rot=45, title='Overall GDP of desired countries')
ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

We can see a steap increase in GDP for USA and China

```python
raw_data_gdp = pd.read_excel("military.xlsx", sheet_name="Share of GDP")
raw_data_gdp.columns = raw_data_gdp.iloc[4]
raw_data_gdp.head()
raw_data_gdp = raw_data_gdp.drop([0,1,2,3,4])
raw_data_gdp = raw_data_gdp.drop(columns = ['Notes'])
raw_data_gdp =  raw_data_gdp.set_index('Country')
country_list = ['USA','China, P.R.', 'Russian Federation', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Israel']
gdp = raw_data_gdp.loc[country_list]
col_list = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]
gdp = gdp[col_list] * 100
gdp_sum = gdp.copy()
gdp_sum = gdp_sum.T.mean()
gdp_sum.to_excel('gdp_mean.xlsx')
```

#### Pie chart for the Military spending as percentage of GDP of each country

```python
#Pie chart for gdp
gdp_sum = gdp.T.mean()
gdp_sum.plot(kind='pie', subplots=True, figsize=(10, 8), autopct='%1.1f%%')
plt.ylabel('Military spending as percentage of GDP of each country')
```

#### We can observe that Saudi Arabia has the highest GDP and then Israel comes in next

### Military spending of Several countries to percentage of GDP year wise

```python
# Plotting Military spending of Several countries to percentage of GDP
# The Pie chart shows the total GDP of the country over the over. However here we can see the GDP in each year
ax1 = gdp.T.plot( kind='bar', width= .6, figsize = (25,15), color = colors, fontsize = 16)
plt.legend(loc = 'best')
plt.title('Military spending of Several countries to percentage of GDP')
ax1.set_xlabel('Year', fontsize = 22)
ax1.set_ylabel('Percentage of GDP', fontsize = 20)
```

#### Correlation of military spending as a percentage of GDP

```python
gdp_df = gdp.T.reset_index()

gdp_df.rename(columns={4: 'Years'}, inplace=True)
melt_data = pd.melt(gdp_df, id_vars=['Years'],var_name='Country')
melt_data.rename(columns={'value':'Military expense as a percentage of GDP'}, inplace=True)
sns.lmplot(x="Years", y="Military expense as a percentage of GDP", hue="Country", data=melt_data, palette="Set2");
```

# Compare the per person military spending to the per person GDP

Importing the gdp per capita (importing data from the world bank dataset which uses SIPRI as data source)

```python
raw_data_gdp_percap = pd.read_csv("gdppercap.csv", skiprows=4)
df_gdp_percap = raw_data_gdp_percap.drop(columns=['Country Code','Indicator Name','Indicator Code','Unnamed: 62'])
df_gdp_percap = df_gdp_percap.rename(columns={'Country Name': 'Country', '2008': 2008, '2009': 2009,'2010': 2010, '2011': 2011,'2012': 2012,'2013': 2013,'2014': 2014,'2015': 2015,'2016': 2016,'2017': 2017})
df_gdp_percap =  df_gdp_percap.set_index('Country')
country_list_gdp_percap = ['United States','China', 'Russian Federation', 'Germany', 'United Kingdom', 'France', 'Italy', 'Saudi Arabia', 'Korea, Rep.', 'Israel']
df_gdp_percap = df_gdp_percap.loc[country_list_gdp_percap]
df_gdp_percap = df_gdp_percap.rename(index={'United States': 'USA', 'China': 'China, P.R.', 'United Kingdom': 'UK','Korea, Rep.': 'Korea, South'})
df_gdp_percap = df_gdp_percap[col_list].round(2)
df_gdp_percap
```

Importing the per capita information from the SIPRI dataset

```python
raw_data_mil_percap = pd.read_excel("military.xlsx", sheet_name="Per capita", skiprows=6)
raw_data_mil_percap.head()
raw_data_mil_percap = raw_data_mil_percap.drop(columns = ['Notes'])
raw_data_mil_percap =  raw_data_mil_percap.set_index('Country')
country_list_percap = ['USA','China, P.R.', 'USSR/Russia', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Israel']
df_percap = raw_data_mil_percap.loc[country_list_percap]
df_percap = df_percap[col_list]
df_percap = df_percap.rename(index={'USSR/Russia': 'Russian Federation'})
df_percap
```

```python
df_percap_mil_gdppercap = (df_percap/df_gdp_percap) * 100
ax2 = df_percap_mil_gdppercap.T.plot.area()
ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5))
df_percap_mil_gdppercap
```

```python
df_percap_mil_gdppercap = (df_percap/df_gdp_percap) * 100
```

```python
df_percap_mil_gdppercap.T.sum().plot(kind='pie', subplots=True, figsize=(10, 8), autopct='%1.1f%%')
plt.ylabel('Per person military spending to the per person GDP')
```

We can observe that a person from Saudi Arabia contributes maximum to the country's military spending, when compared to people from all other countries

#### Year wise comparison of the Per person Military spending of several countries

```python
ax2 = df_percap_mil_gdppercap.T.plot( kind='bar', width= .6, figsize = (25,15), color = colors, fontsize = 16, 
                 title = 'Per person Military spending of Several countries to per person GDP')
plt.legend(loc = 'best')
ax2.set_xlabel('Year', fontsize = 22)
ax2.set_ylabel('Per person Military spending to Per person GDP', fontsize = 20)
```
